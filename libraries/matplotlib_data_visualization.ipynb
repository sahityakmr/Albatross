```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv("res/cars_missing_vals.csv")
cars = cars.rename(columns={'Unnamed: 1':'model'})
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
cars = cars.drop(columns=['S.No'])
df = cars[['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']].corr()
cars.mpg = cars.mpg.astype(float)
cars.head()
```

```python
# see how hp varies with each car
y_hp = cars['hp']
x = range(32)

# to plot bw x and y axis
plt.plot(x, y_hp)
```

```python
y_disp = cars['disp']
plt.plot(x, y_hp)
plt.plot(x, y_disp)

# to show color desc
plt.legend()
```

```python
# stackplot or area-plot
plt.stackplot(x, y_hp, alpha=1, colors='orange')
plt.stackplot(x, y_disp, alpha=0.5, colors='green')
```

```python
# stackplot or area-plot
plt.plot(x, y_hp, linewidth=2.0, color='c')
plt.stackplot(x, y_hp, alpha=0.7, colors='purple')
plt.plot(x, y_disp, linewidth=1.0, color='r')
plt.stackplot(x, y_disp, alpha=0.5, colors='black')
```

```python
# bar chart
x_model = cars['model'].tolist()

# to resize the figure
fig = plt.figure(figsize=(30,15))

# drawing the bar chart
plt.bar(x_model, y_hp, color='gray', alpha=0.9)
```

```python
# drawing the horizontal bar chart
fig = plt.figure(figsize=(17,10))
plt.barh(x_model, y_hp, color='gray', alpha=0.9)
```
